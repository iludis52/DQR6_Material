```python
def validiere_formular(events):
    """
    Simuliert den DFA für die Formularvalidierung mittels if-else-Kette.
    
    :param events: Eine Liste von Strings, die die eingetretenen Events repräsentieren.
    :return: (bool, str) -> (Ist das Formular valide?, Letzter Zustand)
    """
    
    aktueller_zustand = 'q0'
    
    for event in events:
        # --- Zustand q0 ---
        if aktueller_zustand == 'q0':
            if event == 'Event_sub':
                aktueller_zustand = 'q1'
            else:
                return False, aktueller_zustand  # Ungültiges Event
                
        # --- Zustand q1 ---
        elif aktueller_zustand == 'q1':
            if event == 'voll':
                aktueller_zustand = 'q2'
            elif event == 'unvoll':
                aktueller_zustand = 'q0'
            else:
                return False, aktueller_zustand
                
        # --- Zustand q2 ---
        elif aktueller_zustand == 'q2':
            if event == 'mail':
                aktueller_zustand = 'q3'
            elif event == 'nomail':
                aktueller_zustand = 'q0'
            else:
                return False, aktueller_zustand
                
        # --- Zustand q3 ---
        elif aktueller_zustand == 'q3':
            if event == 'visa':
                aktueller_zustand = 'q4'
            elif event == 'novisa':
                aktueller_zustand = 'q0'
            else:
                return False, aktueller_zustand
                
        # --- Zustand q4 ---
        elif aktueller_zustand == 'q4':
            # q4 ist der Endzustand. Jedes weitere Event führt zu einem Fehler,
            # da keine weiteren Übergänge definiert sind.
            return False, aktueller_zustand

    # Überprüfen, ob die Schleife im Endzustand q4 beendet wurde
    ist_akzeptiert = (aktueller_zustand == 'q4')
    
    return ist_akzeptiert, aktueller_zustand
```

```python
# 1. Der "Happy Path" - Alles läuft reibungslos durch
erfolgreicher_durchlauf = ['Event_sub', 'voll', 'mail', 'visa']
valide, end_zustand = validiere_formular(erfolgreicher_durchlauf)
print(f"Test 1 (Happy Path): Valide? {valide} | Endzustand: {end_zustand}")
# Erwartet: True | q4

# 2. Fehlerfall mit Abbruch - Die E-Mail war ungültig
fehler_mail = ['Event_sub', 'voll', 'nomail']
valide, end_zustand = validiere_formular(fehler_mail)
print(f"Test 2 (Fehlerhafte Mail): Valide? {valide} | Endzustand: {end_zustand}")
# Erwartet: False | q0

# 3. Fehler mit anschließendem Neuversuch
neuversuch = ['Event_sub', 'voll', 'nomail', 'Event_sub', 'voll', 'mail', 'visa']
valide, end_zustand = validiere_formular(neuversuch)
print(f"Test 3 (Neuversuch): Valide? {valide} | Endzustand: {end_zustand}")
# Erwartet: True | q4

# 4. Struktureller Fehler - Ungültige Reihenfolge
falsche_reihenfolge = ['Event_sub', 'visa']
valide, end_zustand = validiere_formular(falsche_reihenfolge)
print(f"Test 4 (Falsche Reihenfolge): Valide? {valide} | Endzustand: {end_zustand}")
# Erwartet: False | q1
```
