```python
# Ein 'üblicher' Decision-Tree-Algorithmus wird angewendet. Accuracy von ca, 82.5%. Für das Projekt unwesentlich
import pandas as pd
from matplotlib import pyplot as plt

from sklearn.tree import DecisionTreeClassifier 
from sklearn import tree

from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

m14x14 = pd.read_csv('class_128_14x14.csv')
m14x14.head(6)

feature_spalten = []
for i in range(0, 196):
    feature_spalten.append("p"+str(i))

X = m14x14[feature_spalten]
y = m14x14['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test

clf = DecisionTreeClassifier(random_state=42, max_depth=4)
model = clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)
print("Accuracy:",accuracy_score(y_test, y_pred))

fig = plt.figure(figsize=(20,10))
fig = tree.plot_tree(clf, feature_names=feature_spalten, class_names=['1','2','8'], filled=True, rounded=True, fontsize=10)
```
